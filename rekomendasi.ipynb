Mengambil Product ID skincare yang diminta oleh pengguna

```python
import pandas as pd

age_range = "19 - 24"
skin_type = "Dry"
tag = "cream-1"

from connect import db 
cursor = db.cursor(dictionary=True)
sql = "SELECT products.*, count(reviews.id) as review_count from products join reviews on products.id=reviews.product_id  where tag=%s and reviews.age_range=%s and reviews.skin_type like %s group by products.id"
val = (tag,age_range,f"{skin_type}%")

cursor.execute(sql,val)
data = cursor.fetchall()
data = {row['id']: row for row in data}
pd.DataFrame(data.values()) 
```

Menggabungkan inputan pengguna (product_id, skin_type, age_range)

```python
data_input= []
product_id= list(data.keys())
for id in product_id:
    data_input.append({
        "product_id":id,
        "age_range":age_range,
        "skin_type":skin_type
    })
pd.DataFrame(data_input)
```

Melakukan one hot encoding untuk skin_type dan age_range

```python
data_encoding = []
for inputan in data_input:
    data_encoding.append({
        'product_id': inputan["product_id"],
        'age_range_18 and Under': 1 if inputan["age_range"]== "18 and Under" else 0,
        'age_range_19 - 24': 1 if inputan["age_range"]== "19 - 24" else 0,
        'age_range_25 - 29': 1 if inputan["age_range"]== "25 - 29" else 0,
        'age_range_30 - 34': 1 if inputan["age_range"]== "30 - 34" else 0,
        'age_range_35 - 39': 1 if inputan["age_range"]== "35 - 39" else 0,
        'age_range_40 - 44': 1 if inputan["age_range"]== "40 - 44" else 0,
        'age_range_45 and Above': 1 if inputan["age_range"]== "45 and Above" else 0,
        'skin_type_Combination': 1 if inputan["skin_type"]== "Combination" else 0,
        'skin_type_Dry': 1 if inputan["skin_type"]== "Dry" else 0,
        'skin_type_Normal': 1 if inputan["skin_type"]== "Normal" else 0,
        'skin_type_Oily': 1 if inputan["skin_type"]== "Oily" else 0,
    })
pd.DataFrame(data_encoding)
```

```python
import joblib
import pandas as pd
X = pd.DataFrame(data_encoding)

X
```

Melakukan prediksi rating

```python
dtr1 = joblib.load(f"models/dtr1/{tag}.pkl")
rating = dtr1.predict(X)

df_rating = X.copy()
df_rating['rating'] = rating
df_rating
```

Melakukan prediksi rating text

```python
dtr2 = joblib.load(f"models/dtr2/{tag}.pkl")
rating_text = dtr2.predict(X)

df_rating_text = X.copy()
df_rating_text['rating_text'] = rating_text
df_rating_text
```

Melakukan klasifikasi is_recommended

```python
X2 = pd.DataFrame({
    "rating":rating,
    "rating_text":rating_text,
})

svm = joblib.load(f"models/svm/{tag}.pkl")
is_recommended = svm.predict(X2)

df_is_recommended = X2.copy()
df_is_recommended['is_recommended'] = is_recommended
df_is_recommended
```

Filter yang direkomendasikan dan hitung rata-rata rating

```python
hasil=data.copy()
for id1, rating1, rating_text1, is_recommended1 in zip(product_id, rating, rating_text, is_recommended):
    if is_recommended1 == -1:
        del hasil[id1]
        continue
    hasil[id1].update({
        'rating':rating1,
        'rating_text':rating_text1,
        'average':(rating1+rating_text1)/2,
        'is_recommended':is_recommended1
    })
pd.DataFrame(hasil.values())
```

Menghitung skor akhir menggunakan wilson score interval

```python
import math
def wilson_score(average,review_count):
    p= average/5
    n= review_count
    z= 1.96
    a= p+z**2/(2*n)
    b= math.sqrt((p*(1-p) + z**2/4*n)/n)
    c= 1 + z**2 /n
    atas= (a+b)/c
    bawah= (a-b)/c

    return (atas+bawah)/2
```

```python
for i in hasil.keys():
    hasil[i]['final_score']=0 if hasil[i]['review_count']==0 else wilson_score(hasil[i]['average'], hasil[i]['review_count'])*5
pd.DataFrame(hasil.values())

```

Menampilkan produk yang direkomendasikan dengan menghilangkan produk yang tidak memiliki review dan mengurutkan berdasarkan skor akhir

```python
#hasil_akhir = list(filter(lambda row: row['review_count'] != 0, hasil))
hasil_akhir = [hasil[i] for i in hasil.keys() if hasil[i]['review_count']!= 0]
pd.DataFrame(hasil_akhir)
```

Urutkan berdasarkan final_score tertinggi ke terendah

```python
hasil_akhir.sort(key=lambda row: row['final_score'], reverse=True)
result=pd.DataFrame(hasil_akhir)
pd.DataFrame(result)
```
